```python
import os
from ultralytics import YOLO
```

```python
# change directories to your paths

model = YOLO('C:/Users/britn/Downloads/round3train/weights/best.pt')
image_directory = 'C:/Users/britn/Downloads/subset/images'
results_directory = 'C:/Users/britn/Downloads/subset/results'
os.makedirs(results_directory, exist_ok=True)
```

```python
class_names = {
  0: "Horse",
  1: "bald_eagle",
  2: "black_bear",
  3: "bobcat",
  4: "cheetah",
  5: "cougar",
  6: "deer",
  7: "elk",
  8: "gray_fox",
  9: "hyena",
  10: "lion",
  11: "raccoon",
  12: "red_fox",
  13: "rhino",
  14: "tiger",
  15: "wolf",
  16: "zebra"
}

animal_safety = {
    "Horse" : "Harmless",
    "bald_eagle" : "Harmless",
    "black_bear" : "Harmful",
    "bobcat" : "Harmless",
    "cheetah" : "Harmful",
    "cougar" : "Harmful",
    "deer" : "Harmless",
    "elk" : "Harmless",
    "gray_fox" : "Harmless",
    "hyena" : "Harmful",
    "lion" : "Harmful",
    "raccoon" : "Harmless",
    "red_fox" : "Harmless",
    "rhino" : "Harmful",
    "tiger" : "Harmful",
    "wolf" : "Harmful",
    "zebra" : "Harmless"
}
```

```python
for image_name in os.listdir(image_directory):
    image_path = os.path.join(image_directory, image_name)
    print(f"\nProcessing: {image_name}")

    results = model(image_path, conf = 0.35)
    if isinstance(results, list):
        results = results[0]
        save_img_dir = os.path.join(results_directory, image_name)
        results.save(save_img_dir) 

        labels = results.names 
        boxes = results.boxes.xywh 
        confs = results.boxes.conf  
        classes = results.boxes.cls  

        for result in results:
            for box in result.boxes:
                class_id = int(box.cls)
                class_name = class_names[class_id]
                harmfulness = animal_safety[class_name]
                confidence = box.conf.item()
                print(f"Detected: {class_name}, Confidence: {confidence:.2f}, Status: {harmfulness}")
                print("-----------------------------------")
```
